### Теория графов.
### Алгоритмы на графах.
### Поиск в ширину (BFS, Breadth First Search).

#### Решение задачи acmp 0127 "Путь".

Условие задачи: 

Неориентированный граф на n вершинах задан матрицей смежности. 
Необходимо найти кратчайший путь из вершины start в вершину finish.

План решения задачи:
1. Считиваем n - количество вершин графа.
2. Считываем g - матрицу смежности графа.
3. Считываем вершины номера вершин start и finish.
4. Создадим массив d длины n, в нем будем хранить расстояние от start до остальных вершин графа.  
   Изначально он заполнен -1.
5. Напишем  функцию bfs(x).
6. Запустим поиск в ширину из вершины start, вызвав функцию bfs(start)
7. Выведем на печать d[finish], здесь после отработки bfs хранится ответ на вопрос нашей задачи.





####  Код на языке программирования Python с комментарием.



```python
def bfs(x):                                   # Реализация функции bfs
    d[x] = 0                                  # Ищем расстояние до всех вершин графа от вершины x, d[x] = 0
    q.append(x)                               # Помещаем вершину x в очередь q
    while q:                                  # Пока очередь не пуста, алгоритм продолжает работу
        cur = q.popleft()                     # Извлекаем из головы очереди вершину cur. Она будет текущей. 
        for i in range(n):                    # Находим вершины, смежные с cur
            if g[cur][i] == 1 and d[i] == -1: # Если мы ещё не были в этой вершине, помечаем её как посещенную.
                d[i] = d[cur] + 1             # Меткой будет расстояние от start до смежной с cur вершиной
                q.append(i)                   # Помещаем смежную с cur, ранее не посещенную вершину в очередь q
 
 
from collections import deque                 # Импортируем из библиотеки collections структуру "очередь" deque
 
n = int(input())                              # Считываем количество вершин в графе
g = []                                        # Создаем граф g, изначально пустой
for i in range(n):                            # Считываем граф, заданный матрицей смежности построчно
    g.append(list(map(int, input().split()))) # Добавляем считанную строку в  g

start, finish = map(int, input().split())     # Считываем номера вершин start и finish
start -= 1                                    # Переводим нумерацию вершины из 1-нумерации в 0-нумерацию
finish -= 1                                   # Переводим нумерацию вершины из 1-нумерации в 0-нумерацию
 
q = deque()                                   # Создаем очередь q
d = [-1] * n                                  # Создаем список расстояний d. Изначольно он заполнен -1.
 
bfs(start)                                    # Запускаем поиск в ширину из вершины start
print(d[finish])                              # После отработки алгоритма ответ лежит в d[finish]
```
