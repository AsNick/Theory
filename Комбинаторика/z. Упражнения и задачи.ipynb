### Упражнения и задачи.

**Задача 1.**  
   Учащиеся 9 класса изучают 10 предметов. Сколькими способами  
   можно составить расписание уроков на один день так, чтобы было  
   6 различных уроков?
 

  *Решение:*

$A_{10}^6 = 10 \cdot 9 \cdot 8 \cdot 7 \cdot 6 = 151200$

**Задача 2.**  

   Сколько нужно иметь различных словарей, чтобы непосредственно  
   выполнять переводы с любого из 6 языков (английского, испанского,  
   русского, немецкого, французского, итальянского) на любой другой  
   из этих языков?
   

*Решение:*

$A_6^2 = 6 \cdot 5 = 30$

**Задача 3.**  

   Сколько различных шестизначных чисел можно написать при помощи цифр  
   1, 2, 3, 4, 5, 6, 7, 8, 9? Цифры в записи чисел не повторяются.  
     
     
*Решение:*
  
  $A_9^6 = 9 \cdot 8 \cdot 7 \cdot 6 \cdot 5 \cdot 4 =60480$

**Задача 4.**  
  
   Труппа театра состоит из n актеров. Известно, что четырех претендентов  
   на ведущие роли можно выбрать числом способов, в 56 раз большим, чем  
   выбрать из этой же труппы двух претендентов на главные роли.  
   Сколько артистов в труппе?  
     
*Решение:*  
  
  $A_n^4 = 56 \cdot A_n^2$  
  $n \cdot (n - 1) \cdot (n - 2) \cdot (n - 3) = 56 \cdot n \cdot (n - 1)$  
  $(n - 2) \cdot (n - 3) = 2 \cdot 2 \cdot 2 \cdot 7$  
  $n - 2 = 8$  
  $n - 3 = 7$  
  $n = 10$

**Задача 5.**  
  
   Из 5 чайных чашек, 6 блюдец и 7 чайных ложек хотят накрыть стол для  
   трех человек, дав каждому из них одну чашку, одно блюдце и одну ложку.  
   Сколькими способами можно это сделать?  
     
   *Решение*  
     
$A_5^3 \cdot A_6^3 \cdot A_7^3 = 5 \cdot 4 \cdot 3 \cdot 6 \cdot 5 \cdot 4 \cdot 7 \cdot 6 \cdot 5 = 15120000$
