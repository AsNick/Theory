### Упражнения и задачи.

**Задача 1.**  
   Учащиеся 9 класса изучают 10 предметов. Сколькими способами  
   можно составить расписание уроков на один день так, чтобы было  
   6 различных уроков?
 

  *Решение:*

$A_{10}^6 = 10 \cdot 9 \cdot 8 \cdot 7 \cdot 6 \cdot 5 = 151200$

**Задача 4.**  
  
   Труппа театра состоит из n актеров. Известно, что четырех претендентов  
   на ведущие роли можно выбрать числом способов, в 56 раз большим, чем  
   выбрать из этой же труппы двух претендентов на главные роли.  
   Сколько артистов в труппе?  
     
*Решение:*  
  
  $A_n^4 = 56 \cdot A_n^2$  
  $n \cdot (n - 1) \cdot (n - 2) \cdot (n - 3) = 56 \cdot n \cdot (n - 1)$  
  $(n - 2) \cdot (n - 3) = 2 \cdot 2 \cdot 2 \cdot 7$  
  $n - 2 = 8$  
  $n - 3 = 7$  
  $n = 10$

**Задача 5.**

$A_5^3 \cdot A_6^3 \cdot A_7^3 = 5 \cdot 4 \cdot 3 \cdot 6 \cdot 5 \cdot 4 \cdot 7 \cdot 6 \cdot 5 = 15120000$

#### Задача 6.  
  
  Решите уравнение ($x$ - натуральное число):  
    
а) $\frac{(A_x^5 + A_x^3)}{A_x^3} = 43$  
б) $\frac{(A_x^7 - A_x^5)}{A_x^5} = 89$ 

  *Решение*  
a) $\frac{(A_x^5 + A_x^3)}{A_x^3} = \frac{x \cdot (x - 1) \cdot (x - 2) \cdot (x - 3) \cdot (x - 4) + x \cdot (x - 1) \cdot (x - 2)}{x \cdot (x - 1) \cdot (x - 2)} = (x - 3) \cdot (x - 4) + 1$  
   $(x - 3) \cdot (x - 4) = 42$  
   $(x - 3) \cdot (x - 4) = 2 \cdot 3 \cdot 7$  
   $x - 3 = 7$  
   $x - 4 = 6$  
   x = 10

#### Задача 9.  
  
  На книжной полке стоит собрание сочинений в $20$ томах.  
  Сколькими различнымы способами из можно расставит на полке так, чтобы:  
  а) тома $1$ и $2$ стояли рядом;  
  б) тома $4$ и $5$ не стояли рядом;  
    
   *Решение*
  
 a) У нас есть $2!$ перестановок, состоящих из первого и второго тома.  
 Любую из этих перестановок мы можем разместить 19 способамина полке.
 На оставшихся 18 позиций мы можем разместить 18 томов $18!$ способами.  
 По правилу умножения получим ответ:  
 $19 \cdot P_2 \cdot P_{18} = 19 \cdot 2! \cdot 18!$  
   
 b) *Решение 1.*  
 
   У нас есть $19!$ персестановок, без пятого тома.  
   У нас есть 20 - 2 позиции, куда мы пожем поставить пятый том.
   Тогда по правилу умножения ответ имеет вид:  
   
   $18 \cdot 19!$  
     
   *Решение 2.*  
     
   У нас есть $20!$ способов расставить $20$ томов на полке.  
   В задаче а) мы нашли количество способов расставить 20 томов на полке так, чтобы два помеченных тома не стояли рядом.   
   Вычтем это значение из $20!$, получим ответ на задачу б):  
   
   
   $20! - 19 \cdot 2! \cdot 18!$  
     
   Упростим полученное выражение:  
     
   $20! - 19 \cdot 2! \cdot 18! = 20! - 19! \cdot 2! = 19! \cdot (20 - 2!) = 18 \cdot 19!$ 

#### Задача 10.  
  
  Команда шахматистов состоит из $7$ спортсменов. Перед игрой нужно выбрать шахматиста,  
  выступающего на первой доске и шахматиста, выступающего на второй доске. Остальные  
  $5$ шахматистов произвольным образом рассаживаются за досками с номерами от третьей до седьмой.  
  Сколько имеется вариантов присвоить шахматистам номера от $1$ до $7$?  
    
  *Решение*  
    
  Имеем $A_7^2$ способов выбрать игроков на первую и вторую доску и $P_5$ способов рассадить оставшихся пятерых игроков за досками три - семь.  
  По правилу умножения ответ имеет вид:  
    
   $A_7^2 \cdot P_5 = 7! = 5040$
