### Субфакториал.  
  
#### Задача о проверке контрольных работ.  
  
  $N$ учеников написали контрольную работу. Тетрадки с контрольной работой сданы и лежат на столе учителя. Учитель попросил школьников самостоятельно проверить контрольную. Каждый ученик по очереди подходит к столу учителя и случайным образом выбирает одну тетрадь для проверки.  
  
  а) Сколько существует способов сделать такую выборку?   
  
  Будет нехорошо, если кому-то для проверки попадется его собственная работа.   
  
  б) Сколько существует способов сделать выборку тетрадей так, чтобы ни одному ученику не попалась для проверки своя работа?

Ответ на первый вопрос прост: $\: \: \: \: \: \: \: \: \: \: \: \: N!$  
Ответ на второй вопрос так же прост: $!N$

Рассмотрим перестановку $p$ длины $n$.  
**Неподвижной точкой или порядком** перстановки $p$ называется элемент $p[i]$ такой, что $p[i] = i$.  
**Беспорядком** называется такая перестановка $p$, что для любого элемента $p[i]$ выполняется условие $p[i] \neq i$.  
Другими словами - **беспорядок** это перестановка без неподвижных точек.

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| 5 | 3 | 2 | 4 | 7 | 1 | 6 |  
  
  $p[4] = 4$ - неподвижная точка.

Таким образом, чтобы найти ответ на задачу б), нужно подсчитать количество **беспорядков**  длины $N$.

**Субфакториал** числа $n$ (обозначение: $!n$) определяется как количество беспорядков длины $n$, то есть количество перестановок длины $n$ без неподвижных точек. Название субфакториал происходит из аналогии с факториалом, определяющим общее количество перестановок.

В частности, $!n$ есть число способов положить $n$ писем в $n$ конвертов (по одному в каждый), чтобы ни одно не попало в соответствующий конверт (так называемая «Задача о письмах»).

#### Явная формула субфакториала.

$!n =  n!\left(1-{\frac {1}{1!}}+{\frac{1}{2!}}-{\frac{1}{3!}}+...+(-1)^{n}{\frac{1}{n!}}\right) = n!\sum _{{k=0}}^{n}{\frac  {(-1)^{k}}{k!}}$

#### Реккурентная формула для подсчета субфакториала.

+ ${!n={!(n-1)}\cdot n+(-1)^{n}}$  
+ ${!n=(n-1)\cdot ({!(n-1)}+{!(n-2)})}$ 
+ $!0 = 1$
+ $!1 = 0$  
+ $!2 = 1$
